```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
```

```python
drogon_blue = QuantumCircuit(1,1)
drogon_blue.draw('mpl')
```

```python
drogon_red = QuantumCircuit(1,1)
drogon_red.h(0)
drogon_red.draw('mpl')
```

```python
arrow_blue = QuantumCircuit(1,1)
arrow_blue.measure(0,0)
arrow_blue.draw('mpl')
```

```python
arrow_red = QuantumCircuit(1,1)
arrow_red.h(0)
arrow_red.measure(0,0)
arrow_red.draw('mpl')
```

```python
bb = drogon_blue + arrow_blue
br = drogon_blue + arrow_red
rb = drogon_red + arrow_blue
rr = drogon_red + arrow_red
bb.draw('mpl')
```

```python
qasm_simulator = Aer.get_backend('qasm_simulator')
```

```python
counts_bb = execute(bb,qasm_simulator).result().get_counts()
counts_br = execute(br,qasm_simulator).result().get_counts()
counts_rb = execute(rb,qasm_simulator).result().get_counts()
counts_rr = execute(rr,qasm_simulator).result().get_counts()
```

```python
plot_histogram([counts_bb,counts_br,counts_rb,counts_rr], legend=['bb','rb','br','rr'])
```
